```python
# filling with mean
df["filled"] = df["Sales"].fillna(df["Sales"].mean())
```

```python
df
```

```python
df.rename(columns={"filled": "salesF"}, inplace=True)
```

```python
df
```

```python
df["valueF"] = df["Value"].fillna(df["Value"].mode())
```

```python
df
```

```python
# data aggregation and grouping
grouped_mean = df.groupby("Product")["Value"].mean()
```

```python
grouped_mean
```

```python
grouped_sum = df.groupby(["Product","Region"])["Value"].sum()
```

```python
grouped_sum
```

```python
## aggregate multiple functions

grouped = df.groupby("Region")["Value"].agg(["mean","sum","count"])
grouped
```

## merging and joining DataFrame

```python
d1 = pd.DataFrame({"keys":["A","B","C"],"Value1":[1,2,3]})
d2 = pd.DataFrame({"keys":["A","B","D"],"Value1":[4,5,6]})
```

```python
inner_join = pd.merge(d1,d2,on="keys",how="inner")
inner_join
```

```python
left_join = pd.merge(d1,d2,on="keys",how="left")
left_join
```

```python
right_join = pd.merge(d1,d2,on="keys",how="right")
right_join
```

```python
outer_join = pd.merge(d1,d2,on="keys",how="outer")
outer_join
```

```python
cross_join = pd.merge(d1,d2,on="keys",how="cross")
cross_join
```
