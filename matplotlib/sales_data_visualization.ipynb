# Sales Data Visualization

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("sales_data.csv")
data
```

```python
# plotting total sales by product
tot = data.groupby("Product Category")["Total Revenue"].sum()
tot 
```

```python
tot.plot(kind="bar",color="teal")
```

```python
# sales trend over time
stot = data.groupby("Date")["Total Revenue"].sum().reset_index()
plt.plot(stot["Date"],stot["Total Revenue"])
```
